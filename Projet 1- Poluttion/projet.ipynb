## <h4 style='color:yellow'>Question4 : Analyse descriptive </h4></center>

```python
import pandas as pd
moyenne=data.mean()
mediane=data.median()
ecart_type=data.std()


print(f'La moyenne:\n------------------\n{moyenne}')
print(f'La médiane:\n------------------\n{mediane}')
print(f'L écart_type:\n------------------\n{ecart_type}')

```

```python
Q1=np.percentile(data,25)
Q2=np.percentile(data,50)
Q3=np.percentile(data,75)

print(f'Q1={Q1}, Q2={Q2}, Q3={Q3}')
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]
print(f"Valeurs aberrantes : {outliers}")
```

```python

Q1 = data['CO'].quantile(0.25)
Q2 = data['CO'].quantile(0.50)
Q3 = data['CO'].quantile(0.75)

print(f'Q1={Q1}, Q2={Q2}, Q3={Q3}')
```

## <h4 style='color:yellow'>Question5 : Les corrélations entre les variables et la qualité de l'air </h4></center>

```python
import pandas as pd

pearson = data.corr(method='pearson')
print(f"La corrélation de Pearson :\n-------------------------\n{pearson}\n")
```

```python
if 'Qualite_air' in data.columns:
    target_corr = pearson['Qualite_air']
    print(f" Corrélations avec la variable 'Qualite_Air':\n-------------------------\n{target_corr}\n")
else:
    print("\nLa colonne 'Qualite_air' n'existe pas dans ton fichier. Vérifie son nom exact.")
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(pearson,cmap='coolwarm',annot=True,fmt= '.4f')
plt.title("Matrice de Corrélation - Qualité de l'Air", fontsize=16)
plt.show()
 
```

## <h4 style='color:yellow'>Question6: Visualisations dynamiques  </h4></center>

```python

moyenne = 0
ecart_type = 1

data = np.random.normal(moyenne, ecart_type, 500) 

plt.hist(data, bins=50, color='b', density=True, edgecolor='black')  
plt.title('Histogramme')
plt.xlabel('Valeurs')
plt.ylabel('Densité')


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (np.exp(-((x - moyenne)**2 / (2 * ecart_type**2)))) / (np.sqrt(2 * np.pi) * ecart_type) 

plt.plot(x, p, 'r', linewidth=2)
plt.show()
```
