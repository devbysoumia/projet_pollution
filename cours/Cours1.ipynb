# 9. Les fonctions

```python
def operation(a,b):
    somme=a+b
    produit=a*b
    return somme,produit
```

```python
operation(15,28)
```

```python
s,p= operation(15,28)
print ('La somme est:',s)
print ('Le produit est:',p)    
```

```python
def operation2(a,b):
    somme=a+b
    produit=a*b
    return [somme,produit]
```

```python
List2Result=operation2(15,28)
print(List2Result)  
```

```python
_,p= operation(15,28)
print ('Le produit est:',p)    
```

```python
compteur=0
def incrementer():
   global compteur
   compteur+=1
   print(f"Valeur du compteur à l'interieur: {compteur}")
def affichage():
   print(f"Valeur du compteur à l'exterieur: {compteur}")
```

```python
# fonction récursive
def factorielle_recur(n):
    if n==0:
        return 1
    else:
        return n*factorielle_recur(n-1)
```

```python
print(factorielle_recur(5))
```

```python
print(factorielle_recur(8))
```
