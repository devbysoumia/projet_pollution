# 3. Distribution de la classe

```python

data = pd.DataFrame({
    'class': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C']
})

# Compter le nombre de lignes par classe
count_class = data.groupby('class').size()
print(count_class)
```

# 5. Visualisation 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# histogrammes pour toutes les colonnes numériques
data.hist(bins=20, figsize=(15, 10), layout=(3,3), grid=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(15,20))
plt.show()
```

```python
import matplotlib.pyplot as plt

data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(15,20))
plt.show()
```

```python
#les correlations(examen et projet)
#matrice de correlation

# Corrélation de Pearson
corr = data.corr()

# Corrélation de Spearman
corr2 = data.corr(method='spearman')

# Heatmap pour Pearson
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, fmt='.3f', cmap='pink')
plt.title('Matrice de corrélation')
plt.show()
```
