# 1.avec Numpy


<h7 style='color:aqua'> 1.1 Mesures de tendences centrales </h7>

```python
import numpy as np
from statistics import mode, multimode

x=np.array([6,-1,-5,-7,0,5,6,4,0])
moyenne=np.mean(x)
mediane=np.median(x)
Mode=mode(x)
Multimode=multimode(x)
Multimode=[int(x) for x in Multimode]

print(f'moyenne: {round(moyenne,2)}')
print(f'mediane: {mediane}')
print(f'Mode: {Mode}')
print(f'Multimode: {Multimode}')
```

<h7 style='color:aqua'> 1.2 Mesures de distribution </h7>

```python
varinance=np.var(x)
ecart_type=np.std(x)    
etendu=np.max(x)-np.min(x) #max-min

print(f'varinance: {varinance}')
print(f'écart type: {ecart_type}')
print(f'étendue: {etendu}')
```

```python
from scipy.stats import pearsonr, spearmanr
import numpy as np

x = np.array([-1, 0, 5, 10, 15, 20, 80])
y = np.array([100, 102, 105, 150, 160, 165, 200])
z = np.array([0, -20, -25, -30, -45, -50, -60])
k = np.array([100, -95, 0, 0, -8, 7, 15])

print('---corrélation de Pearson---')
corr_pear, _ = pearsonr(x, y)
print(f'corrélation entre x et y : {corr_pear}')

corr_pear, _ = pearsonr(x, z)
print(f'corrélation entre x et z : {corr_pear}')

corr_pear, _ = pearsonr(x, k)
print(f'corrélation entre x et k : {corr_pear}')

print('---corrélation de Spearman---')
corr_spear, _ = spearmanr(x, y)
print(f'corrélation entre x et y : {corr_spear}')

corr_spear, _ = spearmanr(x, z)
print(f'corrélation entre x et z : {corr_spear}')

corr_spear, _ = spearmanr(x, k)
print(f'corrélation entre x et k : {corr_spear}')
```

# 2.Avec Pandas

```python
import pandas as pd
import numpy as np

x=np.array([5,6,7,8,12,14,24,30])
y=np.array([1,2,3,4,5,6,7,15])
z=np.array([-1,-2,-3,-4,-5,-6,-7,-15])
data={'x':x,'y':y,'z':z}
df=pd.DataFrame(data)

print(df)
```

# 2.1 Mesures de tendences centrales

```python
moyenne=df.mean()
mediane=df.median()
mode_=df.mode()


print(f'La moyenne:\n------------------\n{moyenne}')
print(f'La médiane:\n------------------\n{mediane}')
print(f'Le mode:\n------------------\n{mode_}')
```

# 2.2 Mesures de distribution

```python
varinance=df.var()
ecart_type=df.std() #l'ecart type= racine carré de la variance
etendu=df.max()-df.min() #max-min


print(f'La variance:\n------------------\n{varinance}')
print(f'La écart_type:\n------------------\n{ecart_type}')
print(f'La étendu:\n------------------\n{etendu}') #tous les attrubuts
```

# 2.3 La corrélation

```python
# Corrélation entre les variables d’un DataFrame
import pandas as pd

# Exemple : suppose que df est ton DataFrame
# df = pd.read_csv("ton_fichier.csv")

pearson = df.corr(method='pearson')
print(f"La corrélation de Pearson :\n-------------------------\n{pearson}\n")

spearman = df.corr(method='spearman')
print(f"La corrélation de Spearman :\n-------------------------\n{spearman}\n")

kendall = df.corr(method='kendall')
print(f"La corrélation de Kendall :\n-------------------------\n{kendall}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt  # Correction de l'import

moyenne = 0
ecarttype = 1

data = np.random.normal(moyenne, ecarttype, 500)  # 500 est le nombre de points

# Histogramme
plt.hist(data, bins=50, color='y', density=True, edgecolor='black')  # Ajout de 'data'
plt.title('Histogramme')
plt.xlabel('Valeurs')
plt.ylabel('Densité')

# Courbe de densité gaussienne
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (np.exp(-((x - moyenne)**2 / (2 * ecarttype**2)))) / (np.sqrt(2 * np.pi) * ecarttype)  # Correction de pi et parenthèse

plt.plot(x, p, 'b', linewidth=2)
plt.show()
```

```python
from scipy.stats import shapiro

# Exemple de données
x = [15, -8, 0, 1, 0, 0, 1, -8, 100, 87, 85, 0, 0, 0, 0, 0, 0, 0, 0]

# Test de Shapiro-Wilk
stat, p_value = shapiro(x)

print(f"p-value = {p_value:.4f}")

if p_value > 0.05:
    print("Les données suivent une distribution normale")
else:
    print("Les données ne suivent pas une distribution normale")
```
