# 1. Les Séries


```python
#creer une serie a partir d'une liste
import pandas as pd

data=[-1,8,7,0,3]
serie1=pd.Series(data)
print(serie1)
```

```python
#creer une serie a partir d'un tableau
import pandas as pd
import numpy as np 
data=[21,25,30,42]
tab =np.array(data)
nomLigne=['jan','mars','sep','nov']
serie2=pd.Series(data,index=nomLigne)
print(serie2)
```

```python
#Accés aux données d'une serie
print(serie2['mars'])
print(serie2.values)
print(serie2.index)
print(serie2.describe())
print(f'La moyenne est : {serie2.mean()}')
print(serie2.sep)
print(serie2[serie2>20])

```

```python
serie3=pd.Series([10,20,30,40],index=['A','B','C','D'])
print(serie3)
```

```python
serie31=serie3+5
print(serie31)
```

```python
#appliquer une fonction a une serie
serie32=serie3.apply(lambda x: x**2)
print(serie32)
```

```python
serie321=serie32[serie32>500]
print(serie321)
```

```python
serie4=pd.Series([-10,10,20,-30,40],index=['A','B','C','D','E'])
print(serie4.sum())
print(serie4.mean())
print(serie4.median())
```

```python
#creer une serie avec des valeurs manquantes
serie5=pd.Series([-10,np.nan,20,None,40],index=['A','B','C','D','E'])
print(serie5)
```

```python
#remplacer les valeurs manquantes par 0
serie52=serie5.fillna(0)
print(serie52)
```

```python
#remplacer les valeurs manquantes par la mediane
serie53=serie5.fillna(serie5.median())
print(serie53)
```

```python
#remplacer les valeurs manquantes par l'ecart type
serie54=serie5.fillna(serie5.std())
print(serie54)
```

```python
#supprimer les valeurs manquantes
serie55=serie5.dropna()
print(serie55)
```

```python
#remplacer les valeurs spécifiques
serie6=pd.Series([1,2,3,4],index=['A','B','C','D'])
print(serie6)
print('---apres remplacement---')
#remplacer 1 par 'Un', 2 par 'Deux', 3 par 'Trois', 4 par 'Quatre'
mapping={1:'Un',2:'Deux',3:'Trois',4:'Quatre'}
serie61=serie6.map(mapping)
print(serie61)
```

# 2.DataFrame


```python
import pandas as pd
import numpy as np

list1 = [75, 80, 90]
list2 = [15, 21, 12]
list3 = [100, 98, 100]

# Combiner les listes dans une liste de listes
tab = [list1, list2, list3]
# Définir les noms des colonnes et des lignes
col = ['IA1', 'IA2', 'POC']
ligne = ['Mounia', 'Junior', 'Soumia']
# Créer le DataFrame
df = pd.DataFrame(tab, index=ligne, columns=col)
# Afficher le DataFrame
print(f'dataframe:\n----------------\n{df}')
```

```python
#acceder a une colonne
print(df['IA1'])
```

```python
#acceder a une ligne par son index numérique
print(df.iloc[1])
```

```python
##acceder a une ligne par son index
print(df.loc['Junior'])
```

```python
#filtrer les lignes 
print(df['IA1']>=60)
```

```python
#filtrer les lignes 
print(df[df['IA1']>=60])
```

```python
#ajouter une colonne
df['Logique']=[100,30,100]
print(df)
```
