# 1. Evaluation

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(x_test)

matrice = confusion_matrix(y_test, y_pred)
print(matrice)


def evaluation(matrice):
    TN, FP, FN, TP = matrice.ravel()
    accuracy = (TP + TN) / (TN + FP + FN + TP)
    precision = TP / (TP + FP)
    rappel = TP / (TP + FN)
    f1_score = 2 * precision * rappel / (precision + rappel)
    return TN, FP, FN, TP, accuracy, precision, rappel, f1_score


TN, FP, FN, TP, accuracy, precision, rappel, f1_score = evaluation(matrice)

print("Mesures d'évaluation :")
print("----------------------")
print(f'Accuracy : {round(accuracy * 100, 2)} %')
print(f'Précision : {round(precision * 100, 2)} %')
print(f'Rappel : {round(rappel * 100, 2)} %')
print(f'F1-Score : {round(f1_score * 100, 2)} %')
print(f'TN = {TN}')
print(f'FP = {FP}')
print(f'FN = {FN}')
print(f'TP = {TP}')
```
